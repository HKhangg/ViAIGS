```python
import numpy as np
import pandas as pd
import collections
import math
from pathlib import Path

from tqdm.auto import tqdm

from underthesea import word_tokenize, pos_tag

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report,
    precision_recall_fscore_support,
    f1_score,
    roc_auc_score,
    average_precision_score,
    roc_curve
)

import matplotlib.pyplot as plt
import seaborn as sns

from sentence_transformers import SentenceTransformer

# =========================================================
# FEATURE LIST CUỐI CÙNG
# =========================================================
FEATURES_LIST = [

    # Semantic / lexical diversity
    "entropy_bigram",
    "embedding_norm",
    "top3_word_ratio",

    # Character / stylistic
    "entropy_char",
    "avg_word_length",
    "pos_formality_ratio",
    "long_word_ratio",
    "punct_rate",
    "sentence_len_cv"
]

TEXT_COL = "text"

# =========================================================
# HÀM TÍNH ENTROPY
# =========================================================
def calculate_entropy(tokens):

    if not tokens:
        return 0

    counter = collections.Counter(tokens)

    total = len(tokens)

    return -sum(
        (count / total) * math.log2(count / total)
        for count in counter.values()
    )

# =========================================================
# BIGRAM
# =========================================================
def get_bigrams(tokens):

    return [
        tuple(tokens[i:i+2])
        for i in range(len(tokens)-1)
    ]

# =========================================================
# FEATURE EXTRACTION
# =========================================================
def extract_social_features(text):

    # -----------------------------------------------------
    # EMPTY TEXT
    # -----------------------------------------------------
    if not isinstance(text, str) or text.strip() == "":

        return {
            "entropy_bigram": 0,
            "top3_word_ratio": 0,
            "entropy_char": 0,
            "avg_word_length": 0,
            "pos_formality_ratio": 0,
            "long_word_ratio": 0,
            "punct_rate": 0,
            "sentence_len_cv": 0
        }

    # =====================================================
    # TOKENIZE
    # =====================================================
    words_segmented = word_tokenize(text)

    words_cleaned = [
        w.lower().replace("_", " ")
        for w in words_segmented
        if any(ch.isalnum() for ch in w)
    ]

    chars = [
        c for c in text
        if not c.isspace()
    ]

    bigrams = get_bigrams(words_cleaned)

    # =====================================================
    # BASIC STATS
    # =====================================================
    N = len(words_cleaned)

    # =====================================================
    # ENTROPY
    # =====================================================
    entropy_char = calculate_entropy(chars)

    entropy_bigram = calculate_entropy(bigrams)

    # =====================================================
    # TOP WORD RATIO
    # =====================================================
    if N > 0:

        counter_words = collections.Counter(words_cleaned)

        top3_count = sum(
            count
            for _, count in counter_words.most_common(3)
        )

        top3_word_ratio = top3_count / N

    else:

        top3_word_ratio = 0

    # =====================================================
    # SENTENCE STRUCTURE
    # =====================================================
    sentences = [
        s.strip()
        for s in text
            .replace("?", ".")
            .replace("!", ".")
            .split(".")
        if s.strip()
    ]

    sent_lengths = [
        len(word_tokenize(s))
        for s in sentences
    ]

    sentence_len_cv = (
        np.std(sent_lengths) / np.mean(sent_lengths)
        if sent_lengths and np.mean(sent_lengths) > 0
        else 0
    )

    # =====================================================
    # AVG WORD LENGTH
    # =====================================================
    avg_word_length = (
        np.mean([len(w) for w in words_cleaned])
        if words_cleaned else 0
    )

    # =====================================================
    # LONG WORD RATIO
    # =====================================================
    long_words = [
        w for w in words_cleaned
        if len(w) > 5
    ]

    long_word_ratio = (
        len(long_words) / N
        if N > 0 else 0
    )

    # =====================================================
    # POS FORMALITY
    # =====================================================
    try:

        tags = [
            pos
            for _, pos in pos_tag(text)
        ]

        formal_pos = sum(
            1 for t in tags
            if t in ['N', 'V', 'A', 'P']
        )

        pos_formality_ratio = (
            formal_pos / len(tags)
            if len(tags) > 0 else 0
        )

    except:

        pos_formality_ratio = 0

    # =====================================================
    # PUNCTUATION
    # =====================================================
    puncts = [
        c for c in text
        if c in '.,!?;:"()[]{}'
    ]

    punct_rate = (
        len(puncts) / len(text)
        if len(text) > 0 else 0
    )

    # =====================================================
    # RETURN
    # =====================================================
    return {

        "entropy_bigram": entropy_bigram,

        "top3_word_ratio": top3_word_ratio,

        "entropy_char": entropy_char,

        "avg_word_length": avg_word_length,

        "pos_formality_ratio": pos_formality_ratio,

        "long_word_ratio": long_word_ratio,

        "punct_rate": punct_rate,

        "sentence_len_cv": sentence_len_cv
    }

# =========================================================
# LOAD EMBEDDING MODEL
# =========================================================
embed_model = SentenceTransformer(
    "keepitreal/vietnamese-sbert"
)

# =========================================================
# BATCH EMBEDDING NORM
# =========================================================
def compute_embedding_norm_batch(
    texts,
    batch_size=32
):

    texts = [
        t if isinstance(t, str) else ""
        for t in texts
    ]

    embeddings = embed_model.encode(
        texts,
        batch_size=batch_size,
        show_progress_bar=True,
        normalize_embeddings=False
    )

    return np.linalg.norm(
        embeddings,
        axis=1
    )

# =========================================================
# LOAD OR EXTRACT FEATURES
# =========================================================
def load_or_extract_features(df, metrics_path=None):

    # -----------------------------------------------------
    # CASE 1: LOAD EXISTING FEATURE FILE
    # -----------------------------------------------------
    if metrics_path and Path(metrics_path).exists():

        print(f"\nĐọc feature có sẵn từ: {metrics_path}")

        metrics_df = pd.read_csv(metrics_path)

        # =================================================
        # CHECK MISSING FEATURES
        # =================================================
        missing_cols = [
            f for f in FEATURES_LIST
            if f not in metrics_df.columns
        ]

        # =================================================
        # FULL FEATURE FOUND
        # =================================================
        if len(missing_cols) == 0:

            print("-> Đã tìm thấy đầy đủ feature.")

            return metrics_df[FEATURES_LIST]

        # =================================================
        # MISSING FEATURE
        # =================================================
        print(f"-> Thiếu feature: {missing_cols}")

        print("-> Sẽ tính bổ sung phần còn thiếu.")

        feat_df = metrics_df.copy()

        extracted = pd.DataFrame([
            extract_social_features(t)
            for t in tqdm(
                df[TEXT_COL],
                desc="Extracting missing features"
            )
        ])

        # =================================================
        # ADD MISSING FEATURES
        # =================================================
        for col in missing_cols:

            if col != "embedding_norm":

                feat_df[col] = extracted[col]

        # =================================================
        # EMBEDDING NORM
        # =================================================
        if "embedding_norm" in missing_cols:

            print("Đang tính embedding_norm...")

            feat_df["embedding_norm"] = (
                compute_embedding_norm_batch(
                    df[TEXT_COL].tolist()
                )
            )

        # =================================================
        # SAVE UPDATED FEATURE FILE
        # =================================================
        feat_df.to_csv(
            metrics_path,
            index=False
        )

        print(f"-> Đã cập nhật feature file: {metrics_path}")

        return feat_df[FEATURES_LIST]

    # -----------------------------------------------------
    # CASE 2: CREATE NEW FEATURE FILE
    # -----------------------------------------------------
    print("\nKhông tìm thấy metrics file -> Tính feature mới...")

    feat_df = pd.DataFrame([
        extract_social_features(t)
        for t in tqdm(
            df[TEXT_COL],
            desc="Extracting social features"
        )
    ])

    # =====================================================
    # EMBEDDING NORM
    # =====================================================
    print("Đang tính embedding_norm...")

    feat_df["embedding_norm"] = (
        compute_embedding_norm_batch(
            df[TEXT_COL].tolist()
        )
    )

    # =====================================================
    # SAVE FEATURE FILE
    # =====================================================
    if metrics_path:

        feat_df.to_csv(
            metrics_path,
            index=False
        )

        print(f"-> Đã lưu feature file: {metrics_path}")

    # =====================================================
    # RETURN
    # =====================================================
    return feat_df[FEATURES_LIST]

# =========================================================
# LOAD FEATURE
# =========================================================
df_train_feat = load_or_extract_features(
    df_train,
    "/content/ViAIGS_train_metrics.csv"
)

df_dev_feat = load_or_extract_features(
    df_dev,
    "/content/ViAIGS_dev_metrics.csv"
)

df_test_feat = load_or_extract_features(
    df_test,
    "/content/ViAIGS_test_metrics.csv"
)

# =========================================================
# DATA
# =========================================================
X_train_raw = df_train_feat.fillna(0).values
X_dev_raw = df_dev_feat.fillna(0).values
X_test_raw = df_test_feat.fillna(0).values

y_train = df_train["label"].values
y_dev = df_dev["label"].values
y_test = df_test["label"].values

# =========================================================
# SCALE
# =========================================================
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train_raw)

X_dev = scaler.transform(X_dev_raw)

X_test = scaler.transform(X_test_raw)

# =========================================================
# MODEL
# =========================================================
model = LogisticRegression(
    max_iter=2000,
    C=0.8,
    random_state=42
)

model.fit(X_train, y_train)

# =========================================================
# THRESHOLD TUNING
# =========================================================
print("\n--- Threshold tuning trên DEV ---")

dev_probs = model.predict_proba(X_dev)[:, 1]

best_threshold = 0.5
best_f1 = 0

for th in np.linspace(0.3, 0.7, 41):

    preds = (dev_probs >= th).astype(int)

    f1 = f1_score(
        y_dev,
        preds,
        average='macro'
    )

    if f1 > best_f1:

        best_f1 = f1

        best_threshold = th

print(f"Best threshold = {best_threshold:.2f}")

print(f"Best Macro-F1  = {best_f1:.4f}")

# =========================================================
# MACRO F1 @ 5% FPR
# =========================================================
def macro_f1_at_fpr(
    y_true,
    probs,
    target_fpr=0.05
):

    fpr, tpr, thresholds = roc_curve(
        y_true,
        probs
    )

    idx = np.argmin(
        np.abs(fpr - target_fpr)
    )

    threshold = thresholds[idx]

    preds = (
        probs >= threshold
    ).astype(int)

    macro_f1 = f1_score(
        y_true,
        preds,
        average='macro'
    )

    return macro_f1, threshold

# =========================================================
# EVALUATION
# =========================================================
def evaluate_model(X, y, name):

    probs = model.predict_proba(X)[:, 1]

    preds = (
        probs >= best_threshold
    ).astype(int)

    # =====================================================
    # METRICS
    # =====================================================
    roc_auc = roc_auc_score(
        y,
        probs
    )

    pr_auc = average_precision_score(
        y,
        probs
    )

    f1 = f1_score(
        y,
        preds,
        average='macro'
    )

    macro_f1_5fpr, th_5fpr = macro_f1_at_fpr(
        y,
        probs,
        target_fpr=0.05
    )

    # =====================================================
    # PRINT SUMMARY
    # =====================================================
    print("\n===================================")

    print(name)

    print("===================================")

    print(f"AUC-ROC           : {roc_auc:.4f}")

    print(f"MacroF1 @5%FPR    : {macro_f1_5fpr:.4f}")

    print(f"PR-AUC            : {pr_auc:.4f}")

    print(f"F1-score          : {f1:.4f}")

    print(f"Threshold @5%FPR  : {th_5fpr:.4f}")

    # =====================================================
    # FULL REPORT
    # =====================================================
    print(
        classification_report(
            y,
            preds,
            target_names=[
                'Human (0)',
                'AIG (1)'
            ],
            digits=4
        )
    )

    # =====================================================
    # RETURN RESULT
    # =====================================================
    return {
        "AUC-ROC": roc_auc,
        "MacroF1 @5%FPR": macro_f1_5fpr,
        "PR-AUC": pr_auc,
        "F1-score": f1
    }

# =========================================================
# RUN EVALUATION
# =========================================================
dev_result = evaluate_model(
    X_dev,
    y_dev,
    "DEV"
)

test_result = evaluate_model(
    X_test,
    y_test,
    "TEST"
)

# =========================================================
# FINAL RESULT TABLE
# =========================================================
result_df = pd.DataFrame([
    {
        "Dataset": "DEV",
        **dev_result
    },
    {
        "Dataset": "TEST",
        **test_result
    }
])

print("\n===================================")
print("FINAL RESULT TABLE")
print("===================================")

print(
    result_df.to_string(index=False)
)

# =========================================================
# FEATURE IMPORTANCE
# =========================================================
importance_df = pd.DataFrame({
    'Feature': FEATURES_LIST,
    'Coefficient': model.coef_[0]
})

importance_df = importance_df.sort_values(
    by='Coefficient'
).reset_index(drop=True)

plt.figure(figsize=(10, 6), dpi=120)

colors = [
    '#42a5f5' if c < 0 else '#ef5350'
    for c in importance_df['Coefficient']
]

sns.barplot(
    x='Coefficient',
    y='Feature',
    data=importance_df,
    palette=colors,
    hue='Feature',
    legend=False
)

plt.axvline(
    0,
    color='black',
    linewidth=1
)

plt.title(
    'Feature Importance\n(Blue=Human, Red=AIG)',
    fontsize=14,
    pad=15
)

plt.grid(
    axis='x',
    alpha=0.3
)

plt.tight_layout()

plt.show()

print("\nFeature Importance Table:")

print(
    importance_df.to_string(index=False)
)

# =========================================================
# CORRELATION MATRIX
# =========================================================
corr_matrix = df_train_feat.corr(
    method='spearman'
)

plt.figure(figsize=(10, 8), dpi=120)

mask = np.triu(
    np.ones_like(
        corr_matrix,
        dtype=bool
    )
)

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    square=True
)

plt.title(
    "Feature Correlation Matrix"
)

plt.tight_layout()

plt.show()
```

- Nếu score < threshold → dự đoán là **Human**
